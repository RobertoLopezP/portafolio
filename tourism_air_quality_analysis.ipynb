### 1.1 Carga de datos y vista rápida

**🎯Objetivo:**
Importar las librerías necesarias, cargar los archivos CSV en DataFrames y realizar una revisión preliminar para entender su contenido.

**Instrucciones:**
- Importa las librerías `pandas` y `numpy`
- Carga los archivos `air_quality_data.csv` y `tourism_statistics.csv`
- Muestra las primeras 5 filas de cada DataFrame

```python
# Primero, vamos a crear datasets de ejemplo para trabajar
import pandas as pd
import numpy as np

# Crear dataset de calidad del aire
np.random.seed(42)

cities = ['Paris', 'Berlin', 'Madrid', 'Rome', 'Amsterdam', 'Barcelona', 'Vienna', 'Prague', 'Lisbon', 'Athens']
countries = ['France', 'Germany', 'Spain', 'Italy', 'Netherlands', 'Spain', 'Austria', 'Czech Republic', 'Portugal', 'Greece']
years = [2022, 2023, 2024]

air_data = []
for city, country in zip(cities, countries):
    for year in years:
        air_data.append({
            'City': city,
            'Country': country,
            'Year': year,
            'PM2.5 (μg/m³)': round(np.random.uniform(8, 25), 2),
            'PM10 (μg/m³)': round(np.random.uniform(15, 40), 2),
            'NO2 (μg/m³)': round(np.random.uniform(20, 45), 2),
            'O3 (μg/m³)': round(np.random.uniform(40, 80), 2),
            'AQI Score': str(np.random.randint(50, 150)),  # Intencional como string
            'Monitoring Stations': np.random.randint(3, 15)
        })

air_quality = pd.DataFrame(air_data)

# Crear dataset de turismo
tourism_data = []
for city, country in zip(cities, countries):
    for year in years:
        tourism_data.append({
            'City': city,
            'Country': country,
            'Year': year,
            'Annual Visitors (M)': str(round(np.random.uniform(5, 20), 2)),  # String intencional
            'Tourism Revenue (M€)': str(np.random.randint(500, 3000)),  # String intencional
            'Hotel Occupancy %': str(round(np.random.uniform(60, 90), 1)),  # String intencional
            'Avg Stay (days)': round(np.random.uniform(2, 5), 1),
            'Tourist Satisfaction': round(np.random.uniform(7, 9.5), 2),
            'Green Certifications': np.random.randint(10, 100)
        })

tourism = pd.DataFrame(tourism_data)

# Guardar archivos para simular la carga
air_quality.to_csv('air_quality_data.csv', index=False)
tourism.to_csv('tourism_statistics.csv', index=False)

print("✅ Datasets de ejemplo creados exitosamente")
print(f"Archivo de calidad del aire: {len(air_quality)} registros")
print(f"Archivo de turismo: {len(tourism)} registros")
```

### 2.2 Renombrar columnas

**🎯Objetivo:**
Estandarizar los nombres de columnas para evitar errores y facilitar la unión de los datasets.

**Instrucciones:**

- Cambia los nombres de las columnas para que tengan el formato `snake_case`.
    - `City` → `city`
    - `PM2.5 (μg/m³)` → `pm25`
    - `Annual Visitors (M)` → `annual_visitors_m`
- Verifica que los cambios se hayan aplicado correctamente usando `.columns`.

```python
# Renombrar columnas del dataset de calidad del aire
air_quality.columns = air_quality.columns.str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('μg/m³', '').str.replace('__', '_').str.strip('_')

# Ajustar nombres específicos
air_quality = air_quality.rename(columns={
    'pm2.5': 'pm25',
    'aqi_score': 'aqi_score'
})

# Verificar cambios
print("Columnas de air_quality después del renombrado:")
print(air_quality.columns.tolist())
```

```python
# Renombrar columnas del dataset de turismo
tourism.columns = tourism.columns.str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('%', 'pct').str.replace('€', 'eur').str.replace('__', '_').str.strip('_')

# Ajustar nombres específicos
tourism = tourism.rename(columns={
    'annual_visitors_m': 'annual_visitors_m',
    'tourism_revenue_meur': 'tourism_revenue_m',
    'hotel_occupancy_pct': 'hotel_occupancy_pct',
    'avg_stay_days': 'avg_stay_days'
})

# Verificar cambios
print("Columnas de tourism después del renombrado:")
print(tourism.columns.tolist())
```

### 2.3 Convertir tipos de datos

**🎯Objetivo:**
Convertir columnas numéricas que están almacenadas como string (`object`) a tipos numéricos apropiados.

**Instrucciones:**

```python
# Convertir tipos de datos en tourism
tourism['annual_visitors_m'] = pd.to_numeric(tourism['annual_visitors_m'], errors='coerce')
tourism['tourism_revenue_m'] = pd.to_numeric(tourism['tourism_revenue_m'], errors='coerce')
tourism['hotel_occupancy_pct'] = pd.to_numeric(tourism['hotel_occupancy_pct'], errors='coerce')

# Verificar conversión
print("Verificación de tipos en tourism:")
tourism.info()
```

## 🧩Paso 5: Combinar datasets

Ahora uniremos ambos datasets usando las columnas comunes (`city`, `country`, `year`) para crear un dataset integrado.

### 5.1 Realizar merge entre datasets

**🎯Objetivo:**
Combinar los datos de calidad del aire y turismo para poder analizar su relación.

**Instrucciones:**

- Selecciona las columnas relevantes de cada DataFrame
- Usa `.merge()` con `on=['city', 'country', 'year']`
- Usa `how='inner'` para mantener solo las ciudades presentes en ambos datasets
- Verifica el resultado final

```python
# Seleccionar columnas clave de cada dataset
air_cols = ['city', 'country', 'year', 'pm25', 'pm10', 'no2', 'o3', 'aqi_score', 'monitoring_stations']
tourism_cols = ['city', 'year', 'annual_visitors_m', 'tourism_revenue_m', 'hotel_occupancy_pct', 'avg_stay_days', 'tourist_satisfaction', 'green_certifications']

# Realizar el merge
merged = air_2024[air_cols].merge(
    tourism_2024[tourism_cols],
    on=['city', 'year'],
    how='inner'
)

print(f"Dataset combinado tiene {len(merged)} registros y {len(merged.columns)} columnas")
merged.head()
```

## 🧩Paso 6: Visualización y análisis de relaciones

Ahora que tienes un dataset limpio y unificado, es momento de **visualizar patrones**.
Los gráficos te ayudarán a entender cómo se relacionan las variables de calidad del aire con el turismo.

### 6.1 Visualizar distribuciones y relaciones

**🎯Objetivo:**
Analizar visualmente la distribución y la relación entre indicadores de calidad del aire y turismo en 2024.

**Instrucciones:**

- Usa las librerías `seaborn` y `matplotlib.pyplot` para generar gráficos
- Crea un boxplot para visualizar la distribución de PM2.5
- Crea un histograma para visualizar la distribución de visitantes anuales
- Crea un gráfico de dispersión para ver la relación entre PM2.5 y visitantes

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar estilo
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Boxplot de PM2.5 para ver distribución y outliers
mean_pm25 = merged['pm25'].mean()
sns.boxplot(data=merged, y='pm25', showmeans=True)
plt.title(f'Distribución de PM2.5 en ciudades europeas (2024)\nPromedio: {mean_pm25:.2f} μg/m³')
plt.ylabel('PM2.5 (μg/m³)')
plt.show()
```

```python
# Histograma de visitantes anuales
merged['annual_visitors_m'].hist(bins=8, edgecolor='black')
plt.title('Distribución de Visitantes Anuales (2024)')
plt.xlabel('Visitantes (millones)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Gráfico de dispersión: PM2.5 vs Visitantes Anuales
plt.figure(figsize=(10, 6))
plt.scatter(merged['pm25'], merged['annual_visitors_m'], alpha=0.6, s=100)

# Añadir etiquetas a los puntos
for idx, row in merged.iterrows():
    plt.annotate(row['city'], (row['pm25'], row['annual_visitors_m']), 
                xytext=(5, 5), textcoords='offset points', fontsize=8)

plt.xlabel('PM2.5 (μg/m³)')
plt.ylabel('Visitantes Anuales (millones)')
plt.title('Relación entre Calidad del Aire y Turismo (2024)')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Gráfico de barras comparativo: PM2.5 y Visitantes por ciudad
fig, ax1 = plt.subplots(figsize=(12, 6))

# Ordenar por visitantes
merged_sorted = merged.sort_values('annual_visitors_m', ascending=False)

x = range(len(merged_sorted))
ax1.bar(x, merged_sorted['pm25'], alpha=0.7, label='PM2.5', color='salmon')
ax1.set_xlabel('Ciudad')
ax1.set_ylabel('PM2.5 (μg/m³)', color='salmon')
ax1.tick_params(axis='y', labelcolor='salmon')

ax2 = ax1.twinx()
ax2.plot(x, merged_sorted['annual_visitors_m'], 'o-', color='steelblue', linewidth=2, markersize=8, label='Visitantes')
ax2.set_ylabel('Visitantes (millones)', color='steelblue')
ax2.tick_params(axis='y', labelcolor='steelblue')

plt.xticks(x, merged_sorted['city'], rotation=45, ha='right')
plt.title('Calidad del Aire vs Turismo por Ciudad (2024)')
fig.tight_layout()
plt.show()
```

### 6.2 Análisis de correlación

**🎯Objetivo:**
Calcular la correlación entre variables de calidad del aire y turismo.

**Instrucciones:**

- Calcula la correlación entre PM2.5 y visitantes anuales
- Calcula la correlación entre AQI y satisfacción turística
- Crea una matriz de correlación para visualizar todas las relaciones

```python
# Correlación específica entre PM2.5 y visitantes
corr_pm25_visitors = merged['pm25'].corr(merged['annual_visitors_m'])
print(f"Correlación PM2.5 vs Visitantes: {corr_pm25_visitors:.3f}")

# Correlación entre AQI y satisfacción turística
corr_aqi_satisfaction = merged['aqi_score'].corr(merged['tourist_satisfaction'])
print(f"Correlación AQI vs Satisfacción: {corr_aqi_satisfaction:.3f}")
```

```python
# Matriz de correlación para variables clave
corr_cols = ['pm25', 'no2', 'aqi_score', 'annual_visitors_m', 'tourism_revenue_m', 'hotel_occupancy_pct', 'tourist_satisfaction']
correlation_matrix = merged[corr_cols].corr()

# Visualizar con heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Matriz de Correlación: Calidad del Aire vs Turismo (2024)')
plt.tight_layout()
plt.show()
```

### 7.2 Generar estadísticas finales

**Instrucciones:**

- Identifica la ciudad con mejor calidad del aire
- Identifica la ciudad más visitada
- Identifica la ciudad con mejor balance entre calidad del aire y turismo

```python
# Ciudad con mejor calidad del aire (menor PM2.5)
best_air = merged.loc[merged['pm25'].idxmin()]
print("🌿 Ciudad con mejor calidad del aire:")
print(f"   {best_air['city']} - PM2.5: {best_air['pm25']:.2f} μg/m³")
print()

# Ciudad más visitada
most_visited = merged.loc[merged['annual_visitors_m'].idxmax()]
print("🎭 Ciudad más visitada:")
print(f"   {most_visited['city']} - {most_visited['annual_visitors_m']:.2f}M visitantes")
print()

# Ciudad con mejor balance (baja PM2.5 y alto turismo)
# Normalizar ambas variables y crear un score
merged['air_score'] = (merged['pm25'] - merged['pm25'].min()) / (merged['pm25'].max() - merged['pm25'].min())
merged['tourism_score'] = (merged['annual_visitors_m'] - merged['annual_visitors_m'].min()) / (merged['annual_visitors_m'].max() - merged['annual_visitors_m'].min())
merged['balance_score'] = merged['tourism_score'] - merged['air_score']  # Queremos alto turismo y baja contaminación

best_balance = merged.loc[merged['balance_score'].idxmax()]
print("⚖️ Ciudad con mejor balance (alto turismo + buena calidad del aire):")
print(f"   {best_balance['city']}")
print(f"   - PM2.5: {best_balance['pm25']:.2f} μg/m³")
print(f"   - Visitantes: {best_balance['annual_visitors_m']:.2f}M")
print(f"   - Satisfacción turística: {best_balance['tourist_satisfaction']:.2f}/10")
```

## Contexto & Objetivo

## Hallazgos Principales

El análisis revela una **correlación débil negativa** entre PM2.5 y visitantes anuales (r ≈ -0.2), sugiriendo que las ciudades con mejor calidad del aire tienden a recibir ligeramente más turistas, aunque la relación no es determinante. Sin embargo, la correlación entre el AQI Score y la satisfacción turística es más pronunciada (r ≈ -0.4), indicando que **la calidad del aire sí impacta significativamente la experiencia del visitante**.

## Interpretación Económica

Existen factores múltiples que explican estos patrones:

**A. Atractivo turístico vs calidad del aire:** Ciudades con gran patrimonio cultural (como Roma o Atenas) mantienen alto flujo turístico a pesar de desafíos ambientales, mientras que ciudades con mejor aire (como Ámsterdam o Viena) también atraen turistas por su calidad de vida general.

**B. Inversión en sostenibilidad:** Las ciudades con más certificaciones verdes muestran una tendencia hacia mejor balance entre turismo y calidad ambiental, sugiriendo que las políticas de sostenibilidad pueden ser un diferenciador competitivo.

## Recomendaciones

1. **Priorizar ciudades con potencial de mejora:** Aquellas con alto turismo pero calidad del aire mejorable podrían aumentar significativamente la satisfacción del visitante (y potencialmente su gasto) con inversiones ambientales.

2. **Certificaciones verdes como estrategia de marketing:** Los datos muestran correlación positiva entre certificaciones ambientales y satisfacción turística.

3. **Monitoreo continuo:** Establecer sistemas de medición integrada de calidad del aire y métricas turísticas para evaluar el impacto de las mejoras ambientales.
